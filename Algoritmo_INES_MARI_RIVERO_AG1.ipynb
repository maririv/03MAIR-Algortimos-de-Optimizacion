## Algortimo Torres Hannoi

```python
def torres_hanoi(n, origen, destino, pivote):
  if n == 1:                                            #lo primero que hay que definir es el caso base, es como "parar" la recursividad.
    print(f"Mover bloque desde {origen} a {destino}")
    return

  torres_hanoi(n-1, origen, pivote, destino)
  print(f"Mover bloque desde {origen} a {destino}")

  torres_hanoi(n-1, pivote,destino,origen)
```

```python
torres_hanoi(4,1,3,2)
```

## Algoritmo voraz: Devolver el cambio en monedas:

```python
#Devolver cambio

    #devuelve una lista con cuantas monedas de cada tipo del sistema
    #suponemos que hay infinitos de cada moneda
    #suponemos que está ordenado
def cambio_moneda(cantidad,sistema):  #algoritmo voraz: en cada paso doy el "bocado" más grande
  print(f"El sistema es {sistema}")
  solucion = [0 for i in range(len(sistema))]
  valor_acumulado = 0
  for i in range(len(sistema)):
    moneda = int((cantidad - valor_acumulado)/sistema[i])
    solucion[i] = moneda
    valor_acumulado += moneda*sistema[i]

    if valor_acumulado == cantidad:
      break

  return solucion


```

```python
#Complejidad:
solucion_dic = {}                                                  # 1 asignación
solucion_list = [1000000000]                                       # 1 asignación
def puntos_cercanos(puntos= [14,345,6078,2034,12,34985,3450,45]):
  for i in range(len(puntos)):                                     # n acceso a los datos
    new_list = puntos.copy()                                       # n copias
    new_list.pop(i)                                                # n pop
    for j in range(len(new_list)):                                 # n*(n-1) acceso a los datos
      dist = abs(puntos[i] - new_list[j])                          # n*(n-1) dist
      solucion_dic[(puntos[i], new_list[j])] = dist                # n*(n-1) asignación
      if dist < min(solucion_list):                                # n*(n-1) comparación
        solucion_list.append(dist)                                 # n*(n-1) append
  solucion = min(solucion_list)                                    # 1 asignacion
  for key,value in solucion_dic.items():                           # m acceso a los datos, con m < (n-1)
    if value == solucion:                                          # m comparación, con m < (n-1)
      return key                                                   # 1 return
```

```python
import math


def distancia_minima_puntos(lista):
    lista.sort()
    # Caso base: si la lista tiene solo dos puntos, devuelve la distancia entre ellos
    if len(lista) == 2:
        return abs(lista[1] - lista[0])

    # Caso base: si la lista tiene tres puntos, devuelve la distancia mínima entre ellos
    if len(lista) == 3:
        return min(abs(lista[1] - lista[0]), abs(lista[2] - lista[1]))

    # Divide la lista en dos mitades aproximadamente iguales
    mitad = len(lista) // 2
    mitad_izquierda = lista[:mitad]
    mitad_derecha = lista[mitad:]

    # Encuentra la distancia mínima en cada mitad recursivamente
    distancia_izquierda = distancia_minima_puntos(mitad_izquierda)
    distancia_derecha = distancia_minima_puntos(mitad_derecha)

    # Encuentra la distancia mínima entre las dos mitades
    distancia_minima = min(distancia_izquierda, distancia_derecha)

    # Verifica si hay puntos en las dos mitades que están más cerca entre sí
    distancia_entre_mitades = encontrar_distancia_entre_mitades(lista, distancia_minima)

    # Devuelve la distancia mínima entre las tres distancias encontradas
    return min(distancia_minima, distancia_entre_mitades)


def encontrar_distancia_entre_mitades(lista, distancia_minima):
    # Encuentra el punto medio
    punto_medio = lista[len(lista) // 2]

    # Crea una lista de puntos dentro de la franja central
    franja_central = []
    for punto in lista:
        if abs(punto - punto_medio) < distancia_minima:
            franja_central.append(punto)

    # Ordena la franja central por coordenada y encuentra la distancia mínima
    franja_central.sort()
    minima_distancia_entre_mitades = math.inf
    for i in range(len(franja_central)):
        for j in range(i + 1, len(franja_central)):
            distancia = abs(franja_central[j] - franja_central[i])
            if distancia < minima_distancia_entre_mitades:
                minima_distancia_entre_mitades = distancia

    return minima_distancia_entre_mitades


# Ejemplo de uso
puntos = [4, 8, 1, 2, 9, 5, 11]
distancia_minima = distancia_minima_puntos(puntos)
print("La distancia mínima entre dos puntos es:", distancia_minima)
```

```python
# Calcular la complejidad:

def distancia_minima_puntos(lista):
    lista.sort()                                                                         # 1 ordenar
    # Caso base: si la lista tiene solo dos puntos, devuelve la distancia entre ellos
    if len(lista) == 2:                                                                  # 1 comparación
        return abs(lista[1] - lista[0])                                                  # 1 return

    # Caso base: si la lista tiene tres puntos, devuelve la distancia mínima entre ellos
    if len(lista) == 3:                                                                  # 1 comparación
        return min(abs(lista[1] - lista[0]), abs(lista[2] - lista[1]))                   # 1 return

    # Divide la lista en dos mitades aproximadamente iguales
    mitad = len(lista) // 2                                                              # 1 operacion
    mitad_izquierda = lista[:mitad]                                                      # 1 asignación
    mitad_derecha = lista[mitad:]                                                        # 1 asignación

    # Encuentra la distancia mínima en cada mitad recursivamente
    distancia_izquierda = distancia_minima_puntos(mitad_izquierda)                       # 1 asignación/recursividad
    distancia_derecha = distancia_minima_puntos(mitad_derecha)                           # 1 asignación/recursividad

    # Encuentra la distancia mínima entre las dos mitades
    distancia_minima = min(distancia_izquierda, distancia_derecha)                       # 1 operacion

    # Verifica si hay puntos en las dos mitades que están más cerca entre sí
    distancia_entre_mitades = encontrar_distancia_entre_mitades(lista, distancia_minima) # 1 llamada a función de complejidad (4n^2 - 3n + 6)

    # Devuelve la distancia mínima entre las tres distancias encontradas
    return min(distancia_minima, distancia_entre_mitades)                                # 1 return


def encontrar_distancia_entre_mitades(lista, distancia_minima):
    # Encuentra el punto medio
    punto_medio = lista[len(lista) // 2]                                                 # 1 operación

    # Crea una lista de puntos dentro de la franja central
    franja_central = []                                                                  # 1 asignación
    for punto in lista:                                                                  # n asignaciones
        if abs(punto - punto_medio) < distancia_minima:                                  # n comparaciones
            franja_central.append(punto)                                                 # n operaciones

    # Ordena la franja central por coordenada y encuentra la distancia mínima
    franja_central.sort()                                                                # 1 operación
    minima_distancia_entre_mitades = math.inf                                            # 1 operación
    for i in range(len(franja_central)):                                                 # n asignaciones
        for j in range(i + 1, len(franja_central)):                                      # n(n-1) asignaciones
            distancia = abs(franja_central[j] - franja_central[i])                       # n(n-1) operaciones
            if distancia < minima_distancia_entre_mitades:                               # n(n-1) comparaciones
                minima_distancia_entre_mitades = distancia                               # n(n-1) asignaciones

    return minima_distancia_entre_mitades                                                # 1 return


```

```python
#Aplicar divide y vencerás 2D
import math
import random

def distancia(p1, p2):
    """Calcula la distancia euclidiana entre dos puntos."""
    return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)

def fuerza_bruta(puntos):
    """Encuentra la distancia mínima entre los puntos utilizando fuerza bruta."""
    n = len(puntos)
    min_dist = float('inf')
    for i in range(n):
        for j in range(i + 1, n):
            dist = distancia(puntos[i], puntos[j])
            if dist < min_dist:
                min_dist = dist
    return min_dist

def divide_venceras(puntos):
    n = len(puntos)
    if n <= 3:
        return fuerza_bruta(puntos)

    # Ordenar los puntos por coordenada x
    puntos_ordenados_x = sorted(puntos, key=lambda x: x[0])

    # Dividir el conjunto ordenado en dos mitades
    mitad = n // 2
    puntos_izq = puntos_ordenados_x[:mitad]
    puntos_der = puntos_ordenados_x[mitad:]

    # Encontrar las distancias mínimas en las mitades izquierda y derecha de manera recursiva
    dist_izq = divide_venceras(puntos_izq)
    dist_der = divide_venceras(puntos_der)

    # Encontrar la distancia mínima entre las mitades izquierda y derecha
    dist_min = min(dist_izq, dist_der)

    # Construir la franja central
    franja_central = []
    linea_vertical = puntos_ordenados_x[mitad][0]
    for punto in puntos:
        if abs(punto[0] - linea_vertical) < dist_min:
            franja_central.append(punto)

    # Ordenar los puntos en la franja central por coordenada y
    franja_central_ordenada = sorted(franja_central, key=lambda x: x[1])

    # Verificar distancias mínimas en la franja central
    k = len(franja_central_ordenada)
    for i in range(k):
        param = 2   #calibrado
        for j in range(i + 1, min(i + param, k)):
            dist = distancia(franja_central_ordenada[i], franja_central_ordenada[j])
            if dist < dist_min:
                dist_min = dist

    return dist_min


#puntos_calibrado = [(random.randint(1, 10000), random.randint(1, 10000)) for _ in range(1000)]

# Ejecutar el algoritmo
puntos = [(random.randint(1, 10000), random.randint(1, 10000)) for _ in range(1000)]
distancia_minima = divide_venceras(puntos)

print("La distancia mínima entre los puntos es:", distancia_minima)



```
