## ¿Cuantas posibilidades hay sin tener en cuenta las restricciones?
## Respuesta:
###### El enfoque de este problema es calcular el número de formas de seleccionar 10 actores de entre los 10 disponibles para asignarlos a las 30 tomas, sin restricciones. Esto equivale a contar el número de formas de colocar 10 pelotas en 30 cajas distintas, con repetición y sin importar el orden.
##### La fórmula de combinaciones con repetición, que es:

##### n + r - 1  ! <br>
##### --------------   <br>
#####   r!(n-1)! <br>

##### donde n es el número de tomas (en este caso, 30) y r es el número de actores (en este caso, 10).

##### Aplicando esta fórmula, se tiene:


```
import numpy as np
n = 30
r = 10
posibilidades = np.math.factorial(n + r - 1) / (np.math.factorial(r)*np.math.factorial(n - 1))
print (f"El numero de posibilides sin restricciones es: {posibilidades}")
```

#### Hay varios algoritmos heurísticos que pueden tener una complejidad menor a O(n^2) para ciertos problemas. La complejidad de un algoritmo depende del problema que se esté abordando y de la técnica heurística utilizada.

#### Existen técnicas heurísticas, por ejemplo, algoritmos genéticos, algoritmos de colonias de hormigas, búsqueda tabú y simulación de recocido pueden ser aplicados para resolver problemas complejos con complejidades menores a O(n^2) dependiendo de cómo se implementen y del problema en cuestión.

#### En general, la elección de un algoritmo heurístico depende de la naturaleza del problema y de las restricciones de tiempo y recursos disponibles. Es importante tener en cuenta que los algoritmos heurísticos ofrecen soluciones aproximadas pero no garantizan encontrar la solución óptima. Su fortaleza radica en su eficiencia para resolver problemas difíciles en un tiempo razonable.

```
# Funciones complementarias al algoritmo:

def calcular_coste(solucion):
    """
    Esta función evalúa el coste de utilizar un vecino nuevo como solución
    o quedarse con el actual
    """
    # Inicializar el coste total
    coste = 0
    suma = 0
    # Calcular la suma de los elementos de cada sublista y agregarla al coste total
    for sublista in solucion:
      for sub_sublista in sublista:
        suma += sum(sub_sublista)
    coste += suma

    return coste

def generar_vecindario(solucion_actual):
    """
    Esta función genera una serie aleatoria de vecinos
    """
    vecindario = []

    # Generar vecinos intercambiando dos elementos adyacentes
    i_list  = np.random.randint(1, len(solucion_actual), len(solucion_actual))
    for i in i_list:
        vecino = solucion_actual.copy()
        vecino[i - 1], vecino[i] = vecino[i], vecino[i - 1]
        vecindario.append(vecino)

    return vecindario

def mejor_vecino(solucion_actual):
    """
    Esta función llama a la función generar_vecindario para generar una serie de vecinos, que
    evalúa mediante calcular_coste para quedarse con el mejor (menor coste)
    """
    mejor_vecino = None
    mejor_coste_vecino = float('inf')

    for vecino in generar_vecindario(solucion_actual):
        coste_vecino = calcular_coste(vecino)

        if coste_vecino < mejor_coste_vecino:
            mejor_vecino = vecino
            mejor_coste_vecino = coste_vecino

    return mejor_vecino
```

```
# Definir una lista tabú para almacenar soluciones previamente visitadas
lista_tabu = []                                                                 # 1 asignación

# Definir el número máximo de iteraciones y el tamaño de la lista tabú
iteracion = 0                                                                   # 1 asignación
max_iteraciones = 1000  #calibrar                                               # 1 asignación
tamano_lista_tabu = 5   #calibrar                                               # 1 asignación

# Inicializar la mejor solución
mejor_solucion = poblacion_ini                                                  # 1 asignación

# Comenzar la búsqueda tabú
while iteracion < max_iteraciones:                                              # n comparaciones

    # Generar solución
    nueva_solucion = mejor_vecino(mejor_solucion)                               # n asignaciones

    # Verificar si el vecino está en la lista tabú
    if nueva_solucion not in lista_tabu:                                                # n comparaciones
          mejor_solucion = nueva_solucion                                       # n asignaciones

    # Agregar el movimiento a la lista tabú
    lista_tabu.append(mejor_solucion)                                           # n append

    # Mantener el tamaño de la lista tabú limitado
    if len(lista_tabu) > tamano_lista_tabu:                                     # n comparaciones
        lista_tabu.pop(0)                                                       # n pop

    iteracion +=1                                                               # n asignaciones
```

```
def calcular_coste(solucion):
    """
    Esta función evalúa el coste de utilizar un vecino nuevo como solución
    o quedarse con el actual
    """
    # Inicializar el coste total
    coste = 0                                                                   # 1 asignación
    suma = 0                                                                    # 1 asignación
    # Calcular la suma de los elementos de cada sublista
    # y agregarla al coste total
    for sublista in solucion:                                                   # n accesos
      for sub_sublista in sublista:                                             # n*n accesos
        suma += sum(sub_sublista)                                               # n*n asignación
    coste += suma                                                               # 1 asignación

    return coste

def generar_vecindario(solucion_actual):
    """
    Esta función genera una serie aleatoria de vecinos
    """
    vecindario = []                                                             # 1 asignación

    # Generar vecinos intercambiando dos elementos adyacentes
    i_list  = np.random.randint(1, len(solucion_actual), len(solucion_actual))  # 1 asignación
    for i in i_list:                                                            # n accesos
        vecino = solucion_actual.copy()                                         # n asignación
        vecino[i - 1], vecino[i] = vecino[i], vecino[i - 1]                     # 2n asignación
        vecindario.append(vecino)                                               # n append

    return vecindario                                                           # 1 asignación

def mejor_vecino(solucion_actual):
    """
    Esta función llama a la función generar_vecindario para generar una serie
    de vecinos, que evalúa mediante calcular_coste para quedarse
    con el mejor (menor coste)
    """
    mejor_vecino = None                                                         # 1 asignación
    mejor_coste_vecino = float('inf')                                           # 1 asignación

    for vecino in generar_vecindario(solucion_actual):                          # n accesos
        coste_vecino = calcular_coste(vecino)                                   # n asignación

        if coste_vecino < mejor_coste_vecino:                                   # n comparación
            mejor_vecino = vecino                                               # n asignación
            mejor_coste_vecino = coste_vecino                                   # n asignación

    return mejor_vecino                                                         # 1 asignación
```
